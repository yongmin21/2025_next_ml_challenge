## 모델링

```python
from sklearn.model_selection import train_test_split
train_clean_x = train_clean.drop(columns='clicked')
train_clean_y = train_clean['clicked']
print(train_clean_x)
print(train_clean_y)
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(
    train_clean_x,train_clean_y,
    test_size=0.1,
    random_state=42,
    stratify = train_clean_y
)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_val.shape)
print(Y_val.shape)
```

```python
cols = ['gender', 'age_group', 'inventory_id', 'day_of_week', 'hour'] 

def to_int_df(df, columns):
    for c in columns:
        if c not in df.columns:
            continue
        df[c] = pd.to_numeric(df[c], errors='coerce').fillna(-1).astype('int64')
    return df

X_train = to_int_df(X_train, cols)
X_val   = to_int_df(X_val, cols)
```

```python
test = to_int_df(test,cols)
```

```python
# 3) 모델 파라미터 (GPU + 안정화)
import xgboost as xgb
ratio = 9428464 / 183189  # ≈ 51.5

model = xgb.XGBClassifier(
    objective='binary:logistic',
    eval_metric='auc',
    scale_pos_weight=ratio,
    n_estimators=1000,
    learning_rate=0.05,
    max_depth=8,                 # LightGBM의 num_leaves 대체
    subsample=0.8,
    tree_method='hist',          # GPU: 1) 최신 xgb는 device로 지정
    device='cuda',               # 또는 tree_method='gpu_hist' (구버전)
    predictor='gpu_predictor',
    max_bin=256,                 # 선택(히스토그램 bin)
    min_child_weight=1.0,        # LGB의 min_sum_hessian_in_leaf 대응 개념
    reg_lambda=1.0
)

model.fit(
    X_train, Y_train,
    eval_set=[(X_val, Y_val)],
    verbose=False,
)
```

```python
preds = model.predict_proba(X_val)[:, 1]
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# 1. AUC 점수 계산
auc_score = roc_auc_score(Y_val, preds)
print(f'Validation AUC: {auc_score:.4f}')

# 2. ROC 곡선 시각화
fpr, tpr, thresholds = roc_curve(Y_val, preds)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'AUC = {auc_score:.4f}')
plt.plot([0, 1], [0, 1], 'k--') # 50% 수준의 랜덤 모델
plt.title('ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```
